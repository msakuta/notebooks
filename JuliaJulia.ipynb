Julia set drawn with Julia language :)

$$
z_{n} = {z_{n-1}}^2 + c
$$

```julia
using Plots
```

```julia
function julia(z, c, n)
    ret = zeros(size(z))
    for i = 1:n
        z = z .^ 2 .+ c
        for r = 1:size(z,1)
            for c = 1:size(z,2)
                if ret[r, c] == 0 && 2 < abs(z[r, c])
                    ret[r, c] = i+1
                end
            end
        end
    end
    return ret
end
```

```julia
x_ticks = -2:0.01:2
y_ticks = -2:0.01:2
X = [ j for i=x_ticks, j=y_ticks ]
Y = [ i for i=x_ticks, j=y_ticks ]
julia_result = julia(X + 1im * Y, 0.7 + .3im, 50)
heatmap(julia_result)
```
