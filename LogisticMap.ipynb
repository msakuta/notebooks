# Logistic map

$$
x_{n+1} = rx_n(1-x_n)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rep(x, r):
    return r * x * (1 - x)
```

```python
def calc_logistic(max_x=100, min_r=1., max_r=4., int_r=0.0001):
    x = np.arange(1, max_x)
    r = np.arange(min_r, max_r, int_r)
    y = []
    v = 0.5
    for rv in r:
        for i in x:
            v = rep(v,rv)
        y.append(v)
    return r, y
```

```python
r, y = calc_logistic()
```

```python
plt.plot(r, y, '.', markersize=0.5)
```

```python
r, y = calc_logistic(min_r=3.4, max_r=4, int_r=0.000001)
```

```python
plt.subplots(figsize=(10,10))
plt.plot(r, y, '.', markersize=0.1)
plt.xlim([3.4, 4.])
plt.ylim([0, 1])
```
