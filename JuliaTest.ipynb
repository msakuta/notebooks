```julia
function applyfil(xx)
    x = copy(xx)
    lastval = x[1]
    for i = 1:length(x)
        lastval += 1
        x[i] = min(lastval, x[i])
        lastval = x[i]
    end
    lastval = x[length(x)]
    for i = length(x):-1:1
        lastval += 1
        x[i] = min(lastval, x[i])
        lastval = x[i]
    end
    x
end
```

```julia
@time plot([x, applyfil(x)])
```
