```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as ds
```

```python
all = []
mascal_times = [0.358, 0.329, 0.316, 0.318, 0.311]
all.append(("Mascal\nAST \ninterpreter", np.mean(mascal_times), np.std(mascal_times)))
all.append(("C", 0.004, 0))
all.append(("Rustc", 0.006, 0))
llvm_times = [0.018, 0.016, 0.015, 0.016, 0.018]
llvm_time = np.mean(llvm_times)
all.append(("LLVM", llvm_time, np.std(llvm_times)))
wasm_times = [0.01, 0.008, 0.008, 0.008, 0.009]
wasm_time = np.mean(wasm_times)
all.append(("Wascal", wasm_time, np.std(wasm_times)))
py_times = [0.062, 0.058, 0.062, 0.052, 0.05]
py_time = np.mean(py_times)
all.append(("Python", py_time, np.std(py_times)))
js_times = [0.080, 0.089, 0.076, 0.095, 0.075]
js_time = np.mean(js_times)
all.append(("JavaScript", js_time, np.std(js_times)))
java_times = [0.094, 0.094, 0.089, 0.095, 0.086]
java_time = np.mean(java_times)
all.append(("Java", java_time, np.std(java_times)))
rb_times = [0.131, 0.144, 0.112, 0.133, 0.132]
rb_time = np.mean(rb_times)
all.append(("Ruby", rb_time, np.std(rb_times)))
go_times = [0.113, 0.107, 0.107, 0.109, 0.471]
go_times = [0.118, 0.098, 0.099, 0.101, 0.099]
all.append(("Go run", np.mean(go_times), np.std(go_times)))
gobuild_times = [0.018, 0.019, 0.021, 0.024, 0.022]
gobuild_time = np.mean(gobuild_times)
all.append(("Go build", gobuild_time, np.std(gobuild_times)))
bytecode_times = [0.103, 0.102, 0.105, 0.107, 0.103]
bytecode_times = [0.099, 0.101, 0.099, 0.111, 0.114]
bytecode_time = np.mean(bytecode_times)
all.append(("Mascal\nbytecode", bytecode_time, np.std(bytecode_times)))
bytecode_jump_times = [0.111, 0.110, 0.111, 0.111, 0.159]
bytecode_jump_time = np.mean(bytecode_jump_times)
all.append(("Mascal\nbytecode\nstrflow", bytecode_jump_time, np.std(bytecode_jump_times)))
bytecode_cached_times = [0.110, 0.109, 0.126, 0.111, 0.109]
bytecode_cached_time = np.mean(bytecode_cached_times)
all.append(("Mascal\nbytecode\nstrflow-\ncached", bytecode_cached_time, np.std(bytecode_cached_times)))
all.sort(key=lambda x: x[1], reverse=True)
print(all)
```

```python
names = [entry[0] for entry in all]
times = np.array([entry[1] for entry in all])
errors = np.array([entry[2] for entry in all])
plt.figure(figsize=(12,5))
plt.bar(names, times, yerr=errors, fc="gray")
# plt.errorbar(names, times, yerr=times, fmt=".")
for i, time in enumerate(times):
    plt.text(i, time + 0.01, f"{time:0.3}")
plt.ylabel("Time [s]")
plt.title("Time comparison among languages and runtimes")
plt.tight_layout()
plt.savefig("mandel-time.png")
```
