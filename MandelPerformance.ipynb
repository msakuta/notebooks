```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
llvm_times = [0.018, 0.016, 0.015, 0.016, 0.018]
llvm_time = np.mean(llvm_times)
wasm_times = [0.01, 0.008, 0.008, 0.008, 0.009]
wasm_time = np.mean(wasm_times)
py_times = [0.062, 0.058, 0.062, 0.052, 0.05]
py_time = np.mean(py_times)
js_times = [0.080, 0.089, 0.076, 0.095, 0.075]
js_time = np.mean(js_times)
java_times = [0.094, 0.094, 0.089, 0.095, 0.086]
java_time = np.mean(java_times)
rb_times = [0.131, 0.144, 0.112, 0.133, 0.132]
rb_time = np.mean(rb_times)
go_times = [0.113, 0.107, 0.107, 0.109, 0.471]
go_time = np.mean(go_times)
gobuild_times = [0.018, 0.019, 0.021, 0.024, 0.022]
gobuild_time = np.mean(gobuild_times)
```

```python
names = ["AST \ninterpreter", "Go run", "Ruby", "My own\nBytecode", "Java", "JavaScript", "Python", "Go build", "LLVM IR", "Wasm", "C", "Rustc"]
times = np.array([0.320, go_time, rb_time, 0.105, java_time, js_time, py_time, gobuild_time, llvm_time, wasm_time, 0.004, 0.006 ])
plt.figure(figsize=(10,4))
plt.bar(names, times, fc="gray")
for i, time in enumerate(times):
    plt.text(i, time + 0.01, f"{time:0.3}")
plt.ylabel("Time [s]")
plt.title("Time comparison among languages and runtimes")
plt.tight_layout()
plt.savefig("mandel-time.png")
```
