# Tests of George Marsagria's Xorshift pseudo-random number generator

https://en.wikipedia.org/wiki/Xorshift

```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
```

```python
def xorshift32(state):
    """ Algorithm "xor" from p. 4 of Marsaglia, "Xorshift RNGs" """
    x = state[0];
    x ^= 0xffffffff & (x << 13);
    x ^= 0xffffffff & (x >> 17);
    x ^= 0xffffffff & (x << 5);
    state[0] = x;
    return x;
```

```python
state = [10]
histx, histy = [], []
for _ in range(10000):
    histx.append(xorshift32(state))
    histy.append(xorshift32(state))
plt.plot(histx, histy, '.')
```

```python
state = [1]
for _ in range(10):
    print(xorshift32(state))
```

```python
# The state array must be initialized to not be all zero
def xorshift128(state):
    """ Algorithm "xor128" from p. 5 of Marsaglia, "Xorshift RNGs" """
    t = state[3];
    t ^= 0xffffffff & (t << 11)
    t ^= 0xffffffff & (t >> 8)
    state[3] = state[2]; state[2] = state[1]; state[1] = s = state[0];
    t ^= 0xffffffff & (s)
    t ^= 0xffffffff & (s >> 19)
    state[0] = t;
    return t;
```

```python
state = [10,54,646,13]
histx, histy = [], []
for _ in range(10000):
    histx.append(xorshift128(state))
    histy.append(xorshift128(state))
plt.plot(histx, histy, '.')
```

```python
state = [1,2,3,4]
for _ in range(10):
    print(xorshift128(state))
```
