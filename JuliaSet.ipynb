```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import sys
from matplotlib import rc
```

```python
def julia(z, c, n):
    ret = np.zeros(z.shape)
    for i in range(n):
        z = z ** 2 + c
        ret[(ret == 0) & (2 < np.abs(z))] = i+1
    return ret
```

```python
x_ticks = np.arange(-2,2,0.015)
y_ticks = np.arange(-2,2,0.015)
X, Y = np.meshgrid(x_ticks,y_ticks)
julia_result = julia(X + 1j * Y, 0.7 + .3j, 50)
plt.imshow(julia_result)
plt.colorbar()
print(np.max(julia_result))
plt.figure()
plt.hist(julia_result.flat);
#plt.imshow(Y)
```

```python
x_path = np.arange(0,2*np.pi,0.01*np.pi)
c_path = .6*np.cos(x_path) + 1j * .7*np.sin(0.3 * np.pi + x_path)
plt.plot(c_path.real, c_path.imag, '.-')
```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim((-2, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    #line.set_data([], [])
    return (line,)

# animation function. This is called sequentially
def animate(c):
    julia_result = julia(X + 1j * Y, c, 20)
    ax.clear()
    ax.imshow(julia_result)
    #plt.colorbar()
    sys.stdout.write('.')
    sys.stdout.flush()
    return (line,)

# call the animator. blit=True means only re-draw the parts that 
# have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=c_path, interval=20, blit=True)
```

```python
anim.save('julia.gif', writer='imagemagick', fps=20)
```

```python
mywriter = animation.FFMpegWriter(fps=30)
anim.save('julia.mp4', writer=mywriter)
```
